```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.scatterplot( x = 'age', y = 'fare', data = titanic, hue = 'alive', style = 'class', palette = 'inferno')
plt.title('Titanic Data Analysis')
sns.lineplot( x = 'age', y = 'fare', data = titanic, color = 'green')
```

```python
titanic.head()
```

```python
sns.barplot(x = 'class', y= 'fare', data = titanic, hue = 'sex', palette= 'inferno', estimator = np.max, ci = 60, orient = 'v')
```

```python
sns.barplot(x = 'class', y= 'fare', data = titanic, hue = 'sex', palette= 'inferno', estimator = np.max, errorbar = ('ci', 50), errcolor = 'red', errwidth = 5, saturation = 0.3)
```

```python
flight_df = sns.load_dataset('flights')
```

```python
flight_df.head()
```

```python
flight_df = flight_df.pivot('month', 'year', 'passengers')
```

```python
flight_df.replace('non numeric string', np.nan, inplace=True)
```

```python
flight_df = flight_df.pivot(index = 'month', columns = 'year', values ='passengers')
```

```python
flight_df.head()
```

```python
plt.figure(figsize = (12,6))
```

```python
grid_kws = {"height_ratios": (0.4, 0.05), "hspace": 0.5}
fig, (ax, cbar_ax) = plt.subplots(2, gridspec_kw = grid_kws)
plt.figure(figsize = (12,6))
ax = sns.heatmap(flight_df, cbar_kws={"orientation": "horizontal"}, ax=ax, cbar_ax=cbar_ax)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(titanic.corr())
```
